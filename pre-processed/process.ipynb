```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


print(pd.__version__)
```

```python
data = pd.read_csv('../data/DMSC.csv', encoding='utf-8')

data.iloc[:10, :]
```

```python
print(len(data))
```

```python
print(data['Movie_Name_CN'].value_counts())
```

```python
ls = []
for e in data['Star']:
    ls.append(1 if e > 3 else 0)
print(ls[:10])
data['Label'] = ls
data.iloc[:5, :]
```

```python
drop = ['ID', 'Movie_Name_EN', 'Movie_Name_CN', 'Crawl_Date', 'Number', 'Username', 'Date', 'Star', 'Like']
data = data.drop(drop, axis=1)
data.iloc[:5, :]
```

```python
print(data['Label'].value_counts())
```

```python

print(f"总数据量: {len(data)}")

train_val_df, test_df = train_test_split(
    data,
    test_size=0.1,          
    random_state=42,
    stratify=data['Label']
)


train_df, val_df = train_test_split(
    train_val_df,
    test_size=0.11,
    random_state=42,
    stratify=train_val_df['Label']
)

print(f"训练集数量: {len(train_df)}")
print(f"验证集数量: {len(val_df)}")
print(f"测试集数量: {len(test_df)}")
```

```python
print(type(train_df))
train_df[:5]
```

```python
with open('../data/processed_data.csv', 'w', encoding='utf-8', newline='') as f:
    data.to_csv(f, index=False)

with open('../data/train_data.csv', 'w', encoding='utf-8', newline='') as f:
    train_df.to_csv(f, index=False)

with open('../data/val_data.csv', 'w', encoding='utf-8', newline='') as f:
    val_df.to_csv(f, index=False)

with open('../data/test_data.csv', 'w', encoding='utf-8', newline='') as f:
    test_df.to_csv(f, index=False)
```
